```python
import numpy as np

# Input
X = np.array([[0, 0, 1, 1],
              [0, 1, 0, 1]])

# Target for NOR
t = np.array([[1, 0, 0, 0]])

# Add bias
A = np.vstack((np.ones((1, 4)), X))

# Initialize weights
W = np.array([[0, 0, 0]])

MAX_EPOCH = 10

for ep in range(MAX_EPOCH):
    print(f'\nEpoch {ep+1} ' + '='*40)
    isConv = True
    for x, y in zip(A.T, t.ravel()):
        Z = W @ x.T
        if y > 0 and Z < 0:
            W = W + x
            isConv = False
            print('Update: W + x')
        if y < 1 and Z >= 0:
            W = W - x
            isConv = False
            print('Update: W - x')
        print(f'Input: {x}, Target: {y}, Z: {Z}, Updated W: {W}')
    if isConv:
        print('\nTraining converged!')
        print('Final weights:', W)
        break

# Prediction
Z = W @ A
y_pred = [int(pred >= 0) for pred in Z.ravel()]
print(f'\nFinal prediction: {y_pred}')
```
