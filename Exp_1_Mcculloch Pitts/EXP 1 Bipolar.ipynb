# **Input**

# **AND GATE**

```python
import numpy as np
```

```python
x = np.array([[0,0,1,1], [0,1,0,1]])
print(f'{x.shape=}')
```

# **Parameters**

```python
w = np.array([[1,1]])
print(f'{w.shape=}')
```

```python
theta = 2 
```

**MP NN**

```python
z = w @ x 
print(f'{z=}')
```

```python
y_pred = [int(pred >= theta) for pred in z.ravel()]
print(f'{y_pred = }')
```

# **OR GATE**

```python
theta = 1
```

**MP NN**

```python
z = w @ x 
print(f'{z=}')
```
