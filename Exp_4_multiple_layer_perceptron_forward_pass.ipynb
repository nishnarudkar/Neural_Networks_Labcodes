```python
import numpy as  np
np.set_printoptions(precision=4)
```

```python
W_0 = np.array([[-1,-1,1],
                 [-1,2,1]],dtype=float)
print(f'{W_0.shape=}')
```

```python
X =np.array([[0,-1,1],
             [-1,0,-1]],dtype=float)
print(f'{X.shape=}')
```

```python
f_1 = 'USigmoid'
```

```python
def USigmoid(x):
    return 1/(1+ np.exp(-x))

def BSigmoid(x):
    return (1-np.exp(-x))/(1+np.exp(-x))

def ReLU(x):
    return np.array([max(i,0) for i in x])

def Lin(x):
    return x

def activation(Z,fcn='Lin'):
    A = np.array(list(map(globals()[fcn],Z)))
    return A
```

Forward Pass

```python
if X.shape[0] == W_0.shape[1]:
    print('No bias')
    A_0 = X
else:
    print('Pad bias at top of input')
    A_0 = np.vstack((np.ones((1,X.shape[1])), X))
print(f'{A_0 =}')


print('\n\nInput -> Output')
Z_1 = W_0 @ A_0 
print(f'{Z_1 = }', end='\n\n')
A_1 = activation(Z_1,f_1)
print(f'{A_1 =}')
```
